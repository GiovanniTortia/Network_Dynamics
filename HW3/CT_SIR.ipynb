```python
# Vectors to keep track of ticking times of each node/edge
edge_clocks = np.zeros(len(edges), dtype=float)
node_clocks = np.zeros(n_agents, dtype=float)

# Repeat the simulation N times
for n_sim in range(N):
    # Initialize state vector
    x = np.full(n_agents, 0, dtype=int)
    for i in range(initial_infections):
        x[np.random.randint(n_agents)] = 1

    week = 0
    
    # Initilize ticking times
    for i in range(len(edges)):
        edge_clocks[i] = -np.log(np.random.rand())/beta

    for i in range(n_agents):
        node_clocks[i] = -np.log(np.random.rand())

    # Run simulation until time T_FIN
    while week < T_FIN:
        t_edge = np.min(edge_clocks)
        t_node = np.min(node_clocks)

        # If the first tick is an edge
        if t_edge < t_node:
            # Check if agents get infected
            edge = np.where(edge_clocks == t_edge)[0][0]
            node_a = edges[edge][0]
            node_b = edges[edge][1]

            if x[node_a] == 0 and x[node_b] == 1:
                x[node_a] = 1
                new_infections[week] += 1
            if x[node_a] == 1 and x[node_b] == 0:
                x[node_b] = 1
                new_infections[week] += 1

            # Update ticking time
            edge_clocks[edge] = t_edge - np.log(np.random.rand())/beta

        # Else if the first tick is a node
        else:
            # Check if agent recovers
            node = np.where(node_clocks == t_node)
            if x[node] == 1:
                x[node] = np.random.choice([2, 1], p=[rho, 1-rho])

            # Update ticking time
            node_clocks[node] = t_node - np.log(np.random.rand())

        if t_edge > week and t_node >= week:
            avg_S[week] += len(np.where(x == 0)[0])
            avg_I[week] += len(np.where(x == 1)[0])
            avg_R[week] += len(np.where(x == 2)[0])
            week += 1

avg_S = avg_S/N
avg_I = avg_I/N
avg_R = avg_R/N
new_infections = new_infections/N
```
